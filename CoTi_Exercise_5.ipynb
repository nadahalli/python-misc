# Exercise 3 - Additive Homomorphic ElGamal Encryption 
We are conduting an election with 4 voters voting. There are two candidates, A and B.

```
p = 61318031301932562937212532627060109757384272680362657038289971400896727152243
g = 2

import random

def egcd(a, b):
    if a == 0:
        return (b, 0, 1)
    else:
        g, y, x = egcd(b % a, a)
        return (g, x - (b // a) * y, y)

def modinv(a, m):
    g, x, y = egcd(a, m)
    if g != 1:
        raise Exception('modular inverse does not exist')
    else:
        return x % m

class Key:
    def __init__(self, pri, pub):
        self.pri = pri
        self.pub = pub


def GenKey():
    pri = random.randint(1, p)
    return Key(pri, pow(g, pri, p))

keys = [GenKey() for _ in range (4)]

def ElGamalEncrypt(m, pub):
    x = random.randint(1, p - 1)
    c1 = pow(g, x, p)
    c2 = (m * pow(pub, x, p)) % p
    return c1, c2

def ElGamalDecrypt1(c1, c2, pri):
    s = pow(c1, pri, p)
    return (c2 * modinv(s, p)) % p

def ElGamalDecrypt2(c1, c2, pri):
    return (c2 * pow(c1, (pri * (p - 2)) % p, p)) % p

m = random.randint(1, p-1)
c1, c2 = ElGamalEncrypt(m, keys[0].pub)
print(m == ElGamalDecrypt1(c1, c2, keys[0].pri))
print(m == ElGamalDecrypt2(c1, c2, keys[0].pri))
```
